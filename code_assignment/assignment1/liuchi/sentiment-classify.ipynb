```python
train_data = pd.read_csv("../input/sentiment-analysis-on-movie-reviews/train.tsv.zip",sep = '\t')
test_data = pd.read_csv("../input/sentiment-analysis-on-movie-reviews/test.tsv.zip",sep = '\t')
train_data.head()
```

```python
phrase = train_data['Phrase'].values
sentiment = train_data['Sentiment'].values
```

```python
del train_data
```

```python
phrase[1]
```

```python
#单词出现数量做特征
vectorizer_count = CountVectorizer(stop_words='english',max_features=5000)
corpus = vectorizer_count.fit_transform(phrase)
corpus = corpus.toarray()
print(corpus.shape)
```

```python
X_train,X_test,Y_train,Y_test =train_test_split(corpus,sentiment,test_size=0.2, random_state=42)
print(len(X_train),len(Y_train),len(X_test),len(Y_test))
```

```python
#节省内存
del vectorizer_count
del corpus
```

```python
#model
lr_model = LogisticRegression(solver='liblinear')
lr_model.fit(X_train,Y_train)
accuracy_score(lr_model.predict(X_test),Y_test)
```

```python
del lr_model
del X_train
del Y_train
del X_test
del Y_test
```
